```python
# Calculate IQR
Q1 = df['automated_readability_index'].quantile(0.25)
Q3 = df['automated_readability_index'].quantile(0.75)
IQR = Q3 - Q1
# Identify outliers using IQR method
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_iqr = df[(df['automated_readability_index'] < lower_bound) | (df['automated_readability_index'] > upper_bound)]
print(outliers_iqr.head())
```
